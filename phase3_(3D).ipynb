```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
data = pd.read_csv('electronics_all_data.csv')
fig = plt.figure(figsize=(25, 15))
ax = fig.add_subplot(projection='3d')
X = data['num_reviews']
Y = data['Rating']
Z = data['price']
ax.scatter(X, Y, Z,
           c='indigo',
           s=400,
           alpha=0.5

          )
ax.set_xlabel('Number of Reviews', fontsize=10, labelpad=10)
ax.set_ylabel('Rating (1-5)', fontsize=10, labelpad=10)
ax.set_zlabel('Price ($)', fontsize=10, labelpad=10)
ax.set_title('3D Product Analysis', fontsize=14)
ax.view_init(elev=25, azim=45)
plt.tight_layout()
plt.show()
```
