```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
electronics=pd.read_csv('electronics.csv')
electronics
```

```python
electronics[['price', 'currency']] = electronics['price'].str.split(' ', expand=True)
electronics
```

```python
electronics['price']=pd.to_numeric(electronics['price'])
electronics['price']=electronics['price'].astype(float)
```

```python
electronics=electronics[['title', 'price', 'currency', 'image_url', 'item_id', 'item_url']]
```

```python
electronics['price'].value_counts()
```

```python
electronics.info()
```

```python
electronics.to_csv('electronics_prices.csv', index=False)
```

```python
from scipy.stats import gaussian_kde

electronics_data = pd.read_csv('electronics_prices.csv')
motors_data = pd.read_csv('motors_cleaned.csv')

electronics_prices = electronics_data['price'].values
motors_prices = motors_data['Price'].astype(float).values

price_points = np.linspace(0, 1000, 500)

electronics_distribution = gaussian_kde(electronics_prices)(price_points)
motors_distribution = gaussian_kde(motors_prices)(price_points)

distribution_comparison = np.vstack([electronics_distribution, motors_distribution])

plt.figure(figsize=(12, 4))

plt.imshow(distribution_comparison,
           extent=[0, 1000, 0, 2],
           aspect='auto',
           cmap='viridis',
           origin='lower')

plt.colorbar(label='Product Density')
plt.yticks([0.5, 1.5], ['Electronics', 'Motors'])
plt.xlabel('Price ($)')
plt.title('How Electronics and Motors Products Are Priced')


plt.tight_layout()
plt.show()
```

```python
df_electronics = pd.read_csv('electronics_price_cleaned.csv')
df_motors = pd.read_csv('motors_cleaned.csv')

df_motors['Price']=pd.to_numeric(df_motors['Price'])
df_motors['Price']=df_motors['Price'].astype(float)

df_electronics['price']=pd.to_numeric(df_electronics['price'])
df_electronics['price']=df_electronics['price'].astype(float)

electronics_counts = df_electronics['price'].value_counts().sort_index()
motors_counts = df_motors['Price'].value_counts().sort_index()

plt.figure(figsize=(12, 6))

plt.plot(
    electronics_counts.index,
    electronics_counts.values,
    label='Electronics Prices',
    marker='o',
    linestyle='-',
    color='blue'
)

plt.plot(
    motors_counts.index,
    motors_counts.values,
    label='Motor Prices',
    marker='x',
    linestyle='--',
    color='red'
)

plt.title('Product Count by Price: Electronics vs. Motors')
plt.xlabel('Price ($)')
plt.ylabel('Number of Products')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```

```python
df_electronics = pd.read_csv('electronics_ratings.csv')
df_motors = pd.read_csv('motors_cleaned.csv')

df_motors['Rating']=pd.to_numeric(df_motors['Rating'])
df_motors['Rating']=df_motors['Rating'].astype(float)

df_electronics['Rating']=pd.to_numeric(df_electronics['Rating'])
df_electronics['Rating']=df_electronics['Rating'].astype(float)

electronics_counts = df_electronics['Rating'].value_counts().sort_index()
motors_counts = df_motors['Rating'].value_counts().sort_index()

plt.figure(figsize=(12, 6))

plt.plot(
    electronics_counts.index,
    electronics_counts.values,
    label='Electronics Rating',
    marker='o',
    linestyle='-',
    color='blue'
)

plt.plot(
    motors_counts.index,
    motors_counts.values,
    label='Motor Rating',
    marker='x',
    linestyle='--',
    color='red'
)

plt.title('Product Count by Rating: Electronics vs. Motors')
plt.xlabel('Ratings')
plt.ylabel('Number of Products')
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
