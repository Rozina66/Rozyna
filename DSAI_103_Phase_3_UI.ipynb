```python
import pandas as pd
```

```python
electronics_prices=pd.read_csv('electronics_prices.csv')
electronics_ratings=pd.read_csv('electronics_ratings.csv')
electronics_reviews=pd.read_csv('electronics_reviews_cleaned.csv')
```

```python
motors=pd.read_csv('motors_cleaned.csv')
```

```python
print("Hello!\nThis is Ebay Search engine.\n\nPlease select the category you are searching in\n1 for electronics \n2 for motors\n")

category_num=int(input("Please enter category number to search for: "))
if category_num==1:
    category="electronics"
elif category_num==2:
    category="motors"
else:
    print("Please enter a valid number")

search=input("\nPlease enter a name or brand to search for in "+category+ ":")
q=int(input("\nPlease enter the attribute you would like to search by for " +search+" \n\n1 for price\n2 for rating\n3 for reviews number\n"))
if q==1:
    attribute="price"
elif q==2:
    attribute="rating"
elif q==3:
    attribute="reviews_number"
else:
    print("Please enter a valid number")



if category=="electronics":
    if attribute=="price":
        df=electronics_prices
        result=df[df['title'].str.lower().str.contains(search, na=False)]
        print(result)
    elif attribute=="rating":
        df=electronics_ratings
        if search=="phones" or search=="watches":
          new_df=df[df['Query']==search]
          result=new_df[['Query',"Title","Rating"]]
          print(result)
        else:
            new_df=df[df['Title'].str.lower().str.contains(search, na=False)]
            result=new_df[["Title", "Rating"]]
            print(result)
    elif attribute=="reviews_number":
        df=electronics_reviews
        if search=="phones" or search=="watches" or search=="laptop":
            new_df=df[df['Query']==search]
            result=new_df[['Query',"Title","num_reviews"]]
            print(result)
        else:
          new_df=df[df['Title'].str.lower().str.contains(search, na=False)]
          result=new_df[["Title", "num_reviews"]]
          print(result)


elif category=="motors":
    df=motors
    if attribute=="price":
        if search=="engine" or search=="battery" or search=="fuel cells":
            new_df=df[df['Query']==search]
            result=new_df[['Query',"Title","Price"]]
            print(result)
        else:
            new_df=df[df['Title'].str.lower().str.contains(search, na=False)]
            result=new_df[["Title", "Price"]]
            print(result)

    elif attribute=="rating":
        if search=="engine" or search=="battery" or search=="fuel cells":
            new_df=df[df['Query']==search]
            result=new_df[['Query',"Title","Rating"]]
            print(result)
        else:
            new_df=df[df['Title'].str.lower().str.contains(search, na=False)]
            result=new_df[["Title", "Rating"]]
            print(result)
```
